```python
import pandas as pd
import gdax as g
import matplotlib.pyplot  as plt
```

```python
pub = g.PublicClient()
```

```python
EURHistory = pd.DataFrame(
    pub.get_product_historic_rates('BTC-EUR', granularity =300)
    ,columns=('Time','Low','High','Open','Close','Volume')
)

EURHistory.set_index(['Time'], inplace=True)
EURHistory.set_index(pd.to_datetime(EURHistory.index, unit='s'), inplace=True)

EURHistory['Close'].plot()
plt.show()

EURHistory
```
